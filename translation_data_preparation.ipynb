### Igbo

```python
def split_text_eng(text):
  return text.split("\t")[0]

def split_text_ig(text):
  return text.split("\t")[1]
```

```python
paracrawl_ig=pd.read_csv("/content/drive/MyDrive/translation_data/english-igbo/en-ig.txt",sep="  ", header=None,)
paracrawl_ig["en"]=paracrawl_ig[0].apply(split_text_eng)
paracrawl_ig["ig"]=paracrawl_ig[0].apply(split_text_ig)
paracrawl_ig=paracrawl_ig.reset_index().drop(["index",0],axis=1).drop_duplicates()
```

```python
paracrawl_ig
```

```python
def download_text_from_github(url):
  response = requests.get(url)
  return response.text.split("\n")
```

```python
eng_url="https://raw.githubusercontent.com/IgnatiusEzeani/IGBONLP/master/ig_en_mt/benchmark_dataset/train.en"
ig_url="https://raw.githubusercontent.com/IgnatiusEzeani/IGBONLP/master/ig_en_mt/benchmark_dataset/train.ig"
train_mt=pd.DataFrame({"en":download_text_from_github(eng_url),
                       "ig":download_text_from_github(ig_url)})
```

```python
eng_url="https://raw.githubusercontent.com/IgnatiusEzeani/IGBONLP/master/ig_en_mt/benchmark_dataset/test.en"
ig_url="https://raw.githubusercontent.com/IgnatiusEzeani/IGBONLP/master/ig_en_mt/benchmark_dataset/test.ig"
test_mt=pd.DataFrame({"en":download_text_from_github(eng_url),
                       "ig":download_text_from_github(ig_url)})
```

```python
eng_url="https://raw.githubusercontent.com/IgnatiusEzeani/IGBONLP/master/ig_en_mt/benchmark_dataset/val.en"
ig_url="https://raw.githubusercontent.com/IgnatiusEzeani/IGBONLP/master/ig_en_mt/benchmark_dataset/val.ig"
val_mt=pd.DataFrame({"en":download_text_from_github(eng_url),
                       "ig":download_text_from_github(ig_url)})
```

```python
def create_translation(example,language):
  #print(example[language])
  #print(language["en"])
  return {language:example[language],"en":example["en"]}
```

```python
train_yoruba["translation"]=train_yoruba.apply(create_translation, args=("yo",),axis=1)#.sample(frac=1, random_state=42)
test_yoruba["translation"]=test_yoruba.apply(create_translation, args=("yo",),axis=1)#.sample(frac=1, random_state=42)
test_yoruba["language"]="yo"
train_yoruba["language"]="yo"
train_hausa["translation"]=train_hausa.apply(create_translation, args=("ha",),axis=1)#.sample(frac=1, random_state=42)
test_hausa["translation"]=test_hausa.apply(create_translation, args=("ha",),axis=1)#.sample(frac=1, random_state=42)
test_hausa["language"]="ha"
train_hausa["language"]="ha"
train_igbo["translation"]=train_igbo.apply(create_translation, args=("ig",),axis=1)#.sample(frac=1, random_state=42)
test_igbo["translation"]=test_igbo.apply(create_translation, args=("ig",),axis=1)#.sample(frac=1, random_state=42)
test_igbo["language"]="ig"
train_igbo["language"]="ig"
```
